```python
# samples = wave_data[1] # only keep the left channel
from scipy.io import wavfile
sample_rate, samples = wavfile.read('123.wav')
# print(samples.shape)
# print(raw_samples)
raw_samples = struct.pack("%dh" % len(samples), *samples)
vad = webrtcvad.Vad()
vad.set_mode(3)
window_duration = 0.03
samples_per_window = int(window_duration * sample_rate + 0.4)
bytes_per_sample = 2

segments = []
for start in np.arange(0, len(samples), samples_per_window):
    stop = min(start + samples_per_window, len(samples))
    is_speech = vad.is_speech(raw_samples[start * bytes_per_sample: stop * bytes_per_sample],
                              sample_rate = sample_rate)
#     print(is_speech)
    segments.append(dict(start = start,
                         stop = stop,
                         is_speech = is_speech))
plt.figure(figsize = (10,7))
plt.plot(samples)
ymax = max(samples)

# plot segment identifed as speech
for segment in segments:
    if segment['is_speech']:
        plt.plot([ segment['start'], segment['stop'] - 1], [ymax * 1.1, ymax * 1.1], color = 'orange')
    else:
         plt.plot([ segment['start'], segment['stop'] - 1], [0,0], color = 'orange')
plt.xlabel('sample')
plt.grid()
```

# Case study 2: Spike noise 

```python
def triangle_wave(start,end,period,dense):
    '''
    :param start: the fist value of the wave
    :param end:  the end value of the wave
    :param period:  the period range of the wave
    :param dense:  every period, there are how many points in this period
    :return: the x array and the y array
    '''
 
    xout=[]
    yout=[]
    for i in range(start,end,period):
        x = np.arange(i, i + period,  dense)
        y = np.where(x >= i + period/2, i + period - x, x - i)
 
        xout = np.append(xout, x)
        yout = np.append(yout,y)
    return xout,yout
```

## single spike experiment: 

```python
from scipy.io import wavfile
sample_rate, samples = wavfile.read('123.wav')
# print(samples.shape,max(samples),min(samples))
paras = [10,20,50,100,500]
fig, axs = plt.subplots(5,1,figsize=(20, 20))
ax = 0
for para in paras:
    spike = triangle_wave(0,para,para,1)[1] * 100
    noise = np.zeros_like(samples)
    for i in range(len(noise)):
        if i < len(spike):
            noise[i] = noise[i] + spike[i]
    samples = samples + noise
    raw_samples = struct.pack("%dh" % len(samples), *samples)
    vad = webrtcvad.Vad()
    vad.set_mode(3)
    window_duration = 0.03
    samples_per_window = int(window_duration * sample_rate + 0.4)
    bytes_per_sample = 2

    segments = []
    for start in np.arange(0, len(samples), samples_per_window):
        stop = min(start + samples_per_window, len(samples))
        is_speech = vad.is_speech(raw_samples[start * bytes_per_sample: stop * bytes_per_sample],
                                  sample_rate = sample_rate)
        segments.append(dict(start = start,
                             stop = stop,
                             is_speech = is_speech))
    axs[ax].plot(samples)
    ymax = max(samples)

    # plot segment identifed as speech
    for segment in segments:
        if segment['is_speech']:
            axs[ax].plot([ segment['start'], segment['stop'] - 1], [ymax * 1.1, ymax * 1.1], color = 'orange')
        else:
            axs[ax].plot([ segment['start'], segment['stop'] - 1], [0,0], color = 'orange')
    plt.xlabel('sample')
    axs[ax].grid()
    ax = ax + 1 
```
